```python
#importando biblioteca pandas no Python
import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx") 

#usa o print para ver o dataframe

print(combustiveis_df)

```

```python
#display() mostra o código mais organizado e bonito, podendo substituir o print.

display(combustiveis_df)
```

```python
#Para mostrar as primeiras 5 linhas display(combustiveis_df.head())

display(combustiveis_df.head())
```

```python
#Para exibir as primeiras 15 linhas display(combustiveis_df.head(15))

display(combustiveis_df.head(15))
```

```python
#Comandos Dataframe.shape print(combustiveis_df.shape) mostra o número de linhas e colunas (linhas, colunas)

print(combustiveis_df.shape)
```

```python
#Quais são as colunas e que tipo de dados cada um tem print(combustiveis_df.info())
print(combustiveis_df.info())
```

```python
#A função Dataframe.describe() faz as funções mais básicas que a .shape
print(combustiveis_df.describe())
```

```python
#Filtrar apenas para uma qualquer coluna exemplo REVENDA
display(combustiveis_df['Revenda'])
```

```python
#Aqui criamos um novo dataframe apenas para filtrar as colunas que eu quero.
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']] 
display(ca_df)
```

```python
#Exibe a 4a. linha.
display(ca_df.loc[3])
```

```python
#Exibir da 10a. linha até a 20a.linha
display(ca_df.loc[9:19])
```

```python
#Criar um Dataframe gas_df contendo apenas as 4 colunas(Revenda, Municipio, Produto, Valor de Venda)
#somente com combustivel sendo GASOLINA e exibir na tela.
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

#Como saber o maior preço de venda se chama .max

display(gas_df['Valor de Venda'].max())
```

```python
#Exibir qual é o posto e municipio da gasolina mais cara do pais?
display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
#DataFrame.loc[] com múltiplas condições para filtragem
#Quais são os preços de ETANOL na minha cidade (INDAIATUBA)

etanol_indaiatuba_df = ca_df.loc[(ca_df["Produto"] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
display(etanol_indaiatuba_df)
```

```python
#Ordenaar o preço do menor para o maiorr Dataframe.sort_values (inplace = True ordena e salva)

etanol_indaiatuba_df.sort_values(by = 'Valor de Venda', inplace=True)
display(etanol_indaiatuba_df)
```

```python
#Qual a média de preços dos combustiveis GASOLINA E GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA')
 & (combustiveis_df['Municipio'] == 'SAO PAULO')
  & (combustiveis_df['Produto'].isin(['GASOLINA', 'GASOLINA ADITIVADA'])),
   ['Valor de Venda']].mean())
```

```python
#Como mostrar média de valor de venda por combustível Brasil todo ??

media_por_combustivel_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel_df)
```

```python
#Quero adicionar uma coluna de valor booleano no combutiveis_df chamada "Ativo"
#Que, por Padrão, vai ser True para TODAS as linhas

combustiveis_df['Ativo'] = True
print(combustiveis_df.info())
display(combustiveis_df.head())

```

```python
#E se eu quiser uma coluna nova chamada "Obs" que preencha automaticamente
#"melhor cidade" quando o municipio for Indaiatuba, SP...
############################################
combustiveis_df['Obs'] = "MELHOR CIDADE" 
```

```python
#Exportar para Excel o dataframe com etanol em Indaiatuba...

etanol_indaiatuba_df.to_excel('etanol_indaiatuba.xlsx', sheet_name = 'Etanol em Indaiatuba')
```
